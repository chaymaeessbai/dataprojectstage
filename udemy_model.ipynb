```python
def recommend_course(title, num_of_rec=10):
    # Vérifie que le titre existe dans course_indices
    if title not in course_indices:
        return f"pas de recommandation pour ce cours '{title}'."
    
    # ID pour le titre
    idx = course_indices[title]
    
    # Calcul des scores de similarité
    scores = list(enumerate(cosine_sim_mat[idx]))
    
    # Trier les scores par ordre décroissant
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    # Sélectionner les indices des cours les plus similaires (en excluant le premier car c'est le cours lui-même)
    selected_course_indices = [i[0] for i in sorted_scores[1:num_of_rec+1]]
    selected_course_scores = [i[1] for i in sorted_scores[1:num_of_rec+1]]
    
    # Récupérer les titres des cours correspondants
    result = df['course_title'].iloc[selected_course_indices]
    
    # Créer un DataFrame avec les résultats et les scores de similarité
    rec_df = pd.DataFrame(result)
    rec_df['similarity_scores'] = selected_course_scores
    
    # Retourner le DataFrame des recommandations
    return rec_df


```

```python
# Exemple d'utilisation
recommend_course('Technical Analysis: Understanding Price Action', 30)
```

```python
def recommend_course(title,num_of_rec=10):
    # ID for title
    idx = course_indices[title]
    # Course Indice
    # Search inside cosine_sim_mat
    scores = list(enumerate(cosine_sim_mat[idx]))
    # Scores
    # Sort Scores
    sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)
    # Recomm
    selected_course_indices = [i[0] for i in sorted_scores[1:]]
    selected_course_scores = [i[1] for i in sorted_scores[1:]]
    result = df['course_title'].iloc[selected_course_indices]
    rec_df = pd.DataFrame(result)
    rec_df['similarity_scores'] = selected_course_scores
    return rec_df.head(num_of_rec)
```
